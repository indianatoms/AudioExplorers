```python
import matplotlib.pyplot as plot
from scipy.io import wavfile
import numpy as np
import pandas as pd
```

```python

 

# Read the wav file (mono)

path = ['music_0.wav',  'music_2.wav',  'music_4.wav',  'other_1.wav',  'other_3.wav',
'music_1.wav', 'music_3.wav',  'other_0.wav' , 'other_2.wav',  'other_4.wav'
]


for x in path:
    samplingFrequency, signalData = wavfile.read('../Audio examples/' + x)


    # Plot the signal read from wav file

    plot.subplot(211)

    plot.title('Spectrogram of a wav file with ' + x)

    

    plot.plot(signalData)

    plot.xlabel('Sample')

    plot.ylabel('Amplitude')

    

    plot.subplot(212)

    plot.specgram(signalData,Fs=samplingFrequency)

    plot.xlabel('Time')

    plot.ylabel('Frequency')

    

    plot.show()
```
